# Dividen Data
## Load Data

The resulting data set is date and closing mean of the 10 companies.

Load the data from csv file. Each record has these values:    

1. date
2. Open Price 

**Predictor Variable:**
3. Close Price   

4. High Price  
5. Low Price                          
6. Volume
7. Name


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df1 = pd.read_csv('./dividends/ADM_data.csv', sep=',',header=0)
df1.head()
```

```python
df2 = pd.read_csv('./dividends/BEN_data.csv', sep=',',header=0)
df2.head()
```

```python
df3 = pd.read_csv('./dividends/EMR_data.csv', sep=',',header=0)
df3.head()
```

```python
df4 = pd.read_csv('./dividends/GPC_data.csv', sep=',',header=0)
df4.head()
```

```python
df5 = pd.read_csv('./dividends/GWW_data.csv', sep=',',header=0)
df5.head()
```

```python
df6 = pd.read_csv('./dividends/KMB_data.csv', sep=',',header=0)
df6.head()
```

```python
df7 = pd.read_csv('./dividends/KO_data.csv', sep=',',header=0)
df7.head()
```

```python
df8 = pd.read_csv('./dividends/LEG_data.csv', sep=',',header=0)
df8.head()
```

```python
df9 = pd.read_csv('./dividends/TGT_data.csv', sep=',',header=0)
df9.head()
```

```python
df10 = pd.read_csv('./dividends/WBA_data.csv', sep=',',header=0)
df10.head()
```

```python
dfs = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10]
dfs_concat = pd.concat(dfs)
dfs_concat.head()
```

```python
dfs_concat.tail()
```

```python
dfs_concat = dfs_concat.groupby('date')[['close']].mean()
dfs_concat.head()
```

```python
dfs_concat.to_csv('./DividenMean.csv')
```
